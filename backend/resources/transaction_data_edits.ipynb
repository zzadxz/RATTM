```python
import random
from datetime import datetime, timedelta

# generate a random timestamp between start and end (dates)
def generate_random_timestamp(start, end):
    delta = end - start
    random_seconds = random.randint(0, int(delta.total_seconds()))
    random_date = start + timedelta(seconds=random_seconds)
    # ISO 8601 string
    return random_date.strftime("%Y-%jT%H:%M:%S.%fZ")  # Use %j for day of the year
```
